```python
import os
# 리턴 값 = os.path.exists(파일명)
binsrc = ''
binsrc_encode = binsrc.encode()
binsrc_encode
```

```python
if os.path.exists(binsrc):
    dfp = open.(binsrc, 'rb')
    while True:
        sbyte = dfp.read()
        if not sbyte:
            break;
        dfp.write()
```
